Load dữ liệu lên

```python
import pandas as pd
file = ('/content/Full_Mark_2020.csv')
df = pd.read_csv(file)
df.columns= ['No','Code','Diali','GDCD','Hoahoc','KHTN','KHXH','LichSu','Ngoaingu','NguVan','Sinhhoc','Toan','Vatli','city']
df.info()
```

Hai dòng đầu tiên chuyển đổi cột 'Vatli' và 'city' sang kiểu dữ liệu số. Nếu có lỗi, nó sẽ thay thế giá trị lỗi bằng NaN. df.info() sau đó được gọi để hiển thị thông tin tóm tắt về DataFrame df sau khi đã chuyển đổi kiểu dữ liệu của cột 'Vatli' và 'city'.

```python
df['Vatli'] =pd.to_numeric(df['Vatli'], errors='coerce')
df['city'] = pd.to_numeric(df['city'], errors='coerce')

df.info()
```

```python
#Kiểm tra dữ liệu trùng lặp
duplicate_rows = df[df.duplicated(subset='Code')]
print('Số lượng dữ liệu trùng lâp: ' ,len(duplicate_rows))
```

```python
df = df.drop_duplicates(subset='Code')
df

```

**CÂU 2: Số bài thi môn toán phổ điểm bài thi của môn toán **

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Tạo bins
bins = pd.interval_range(start=0, end=10, freq=0.2, closed='left')

# Chia điểm Toán vào bins và đếm số lượng
pho_diem_toan = pd.cut(df['Toan'], bins=bins).value_counts().sort_index()

# Vẽ biểu đồ cột
fig, ax = plt.subplots(figsize=(10, 6))  # Điều chỉnh kích thước đồ thị
bars = ax.bar(np.arange(len(pho_diem_toan)), pho_diem_toan.values, width=0.8, color='skyblue', edgecolor='black')  # Điều chỉnh màu sắc và viền cột
ax.set_xlabel("Điểm số", fontsize=12)
ax.set_ylabel("Số lượng bài thi", fontsize=12)
ax.set_title("Phổ điểm bài thi môn Toán", fontsize=14)

ax.set_xticks(np.arange(len(pho_diem_toan)))
ax.set_xticklabels(pho_diem_toan.index.astype(str), rotation=45, ha='right', fontsize=10)  # Điều chỉnh nhãn trục x
ax.grid(axis='y', linestyle='--', alpha=0.7)  # Thêm lưới mờ
# Làm tròn điểm số trên trục x
labels = [f'{interval.left:.1f}-{interval.right:.1f}' for interval in pho_diem_toan.index]
ax.set_xticklabels(labels, rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```

Tính điểm trung bình điểm trung vị độ lệch chuẩn

```python
import pandas as pd

# Tính toán các giá trị thống kê
diem_trung_binh = df['Toan'].mean()
diem_trung_vi = df['Toan'].median()
do_lech_chuan = df['Toan'].std()

# Tạo DataFrame tổng hợp
data = {'Giá trị': ['Điểm trung bình', 'Điểm trung vị', 'Độ lệch chuẩn'],
        'Toán': [diem_trung_binh, diem_trung_vi, do_lech_chuan]}
bang_tong_hop = pd.DataFrame(data)

# Hiển thị bảng
bang_tong_hop
```
